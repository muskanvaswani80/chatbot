```python
#  shopping_chatbot.py

import time
from IPython.display import display, HTML

def welcome():
    print("Done by Muskan Vaswani\n")
    print(" Welcome to ShopEasy! My name is Brownie...... I'm your shopping assistant.")
    print(" How can I help you today?")
    print(" You can type 'browse', 'order', 'track', 'help', 'return' or 'exit'.")

def show_products():
    print(" Here are our top products:")
    print("1. Apple iPhone 14 - $799")
    print("2. formal wear for men and women 50-5000$")
    print("3. Cartier watches 1000-10,000$")
    print("4. Nike Sneakers - $129")

def place_order():
    show_products()
    choice = input("\n Enter the number of the product you'd like to order: ")
    if choice in ['1', '2', '3', '4']:
        print(" Great! Your order has been placed. You'll receive a confirmation email soon.")
    else:
        print(" \nSorry, I didn't understand that. Please try again.")

def track_order():
    order_id = input(" Please enter your order ID: ")
    print(f" Tracking order #{order_id}...")
    time.sleep(2)
    print(" You'll will recieve your parsel in 2 to 3 business days!!")



def return_order():
    order_id = input("\n Please enter the order ID for the item you wish to return: ")
    print(f" Initiating a video call to verify the quality of the product for order #{order_id}...")
    time.sleep(3)
    print(" \nConnecting to a support agent...")
    time.sleep(2)
    # Display a simple HTML representation of a video call screen with icons
    video_call_html = """
    <div style="border: 1px solid black; width: 400px; height: 300px; text-align: center; display: flex; flex-direction: column; justify-content: center; align-items: center;">
        <div style="font-size: 24px; margin-bottom: 20px;">Video Call Screen (Agent View)</div>
        <div style="font-size: 40px;">
            <span style="margin-right: 20px; color: red;">&#x274C;</span> <!-- Red X (Decline) -->
            <span style="color: green;">&#x2705;</span> <!-- Green Check (Answer) -->
        </div>
    </div>
    """
    display(HTML(video_call_html))
    time.sleep(5)
    print(" \nThank you. Your return request has been approved. You will receive a return label via email.")


def help_info():
    print("""\n
 I'm here to help! Here's what you can do:
- Type 'browse' to view products
- Type 'order' to place a new order
- Type 'track' to track your order
- Type 'return' to return your order
- Type 'exit' to end this chat
""")

def chatbot():
    welcome()
    while True:
        user_input = input("\nYou: ").lower()

        if user_input in ['browse', 'products', 'show products']:
            show_products()
        elif user_input in ['order', 'buy']:
            place_order()
        elif user_input in ['track', 'track order']:
            track_order()
        elif user_input in ['help', 'info']:
            help_info()
        elif user_input == 'return':
            return_order()
        elif user_input in ['exit', 'quit', 'bye']:
            print("Thanks for visiting ShopEasy. Have a great day!")
            break
        else:
            print("\nSorry, I didn't understand that. Type 'help' for options.")

if __name__ == "__main__":
    chatbot()
```
